# **Pandas**

[__Pandas__](https://pandas.pydata.org/docs/) adalah _library_ pada bahasa Python yang menyediakan struktur data dan fungsi-fungsi yang mempermudah pemrosesan dan analisis data. ## ***Series***

Dalam *library* Pandas, terdapat [***series***](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html) yang digunakan untuk menyimpan koleksi data **satu** dimensi, mirip dengan *array* pada bahasa pemrograman umumnya. *Series* juga memiliki index yang dimulai dari 0 hingga N-1 dengan N adalah jumlah elemen pada *series* tersebut sehingga memampukan kita untuk melakkuan *indexing* dan *slicing* data. Sama seperti array, *series* dapat menyimpan elemen dengan berbagai jenis data, seperti *integer*, *string*, *float*, dan lainnya.

Salah satu cara yang umum digunakan untuk membuat objek **Series** adalah sebagai berikut:
```
pd.Series(data)
```
Di mana `data` dapat berupa list, tuple, atau dictionary.

````python
# Contoh pembuatan objek Series dari sebuah tuple
series = pd.Series((10, 20, 30, 40, 50))
series
````

Saat *series* di atas ditampilkan sebagai output, terdapat dua buah kolom: kolom sebelah kiri merupakan `index` dan kolom sebelah kanannya adalah `values` yang berkorespondensi satu sama lain.

````python
# Contoh melakukan indexing pada objek Series
series[2]
````

````python
# Contoh melakukan slicing pada objek Series
series[:3]
````

## ***DataFrame***

Untuk menyimpan data yang berdimensi dua seperti tabel, kita bisa menggunakan objek [***DataFrame***](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html). Setiap data disimpan pada pasangan *index* (baris) dan *column* (kolom) yang unik. Setiap *column* pada *DataFrame* dapat menyimpan suatu tipe data yang berbeda-beda seperti *integer*, *string*, atau *float* sehingga memungkinkan menyimpan data yang variatif.

Untuk membuat objek *DataFrame*, kita bisa menggunakan data dari berbagai struktur data dan sumber file, seperti kumpulan list, tuple, *Series*, CSV, Excel, dan SQL. *Syntax* umum untuk membuat objek DataFrame adalah sebagai berikut:

```
pd.DataFrame(data, columns=column_names)
```

`data` dapat berupa list, tuple, dictionary, atau *Series*
`column_names` berupa list yang berisi daftar nama kolom yang berurutan. Jika tidak diberikan, nama kolom akan dihasilkan secara otomatis sebagai integer mulai dari 0

### Membuat *DataFrame* dari *Series*

````python
df = pd.DataFrame(series, columns=["Buah"])
df
````

### Mengakses *Index* (Baris) dan *Column* pada *DataFrame*

Terdapat beberapa cara untuk mengakses suatu atau sekumpulan data pada DataFrame dengan memanfaatkan informasi *index* dan *column*:

#### Mengakses *Column* dengan *Square Bracket*

Kita dapat mengakses satu atau beberapa column menggunakan *syntax square bracket* (`[]`), mirip seperti pada *List* atau *Series*. Apabila hanya menggunakan satu square bracket, maka yang dihasilkan adalah sebuah Series tetapi hanya bisa menyimpan satu column. Dengan dua square bracket, kita bisa menghasilkan DataFrame baru.

````python
# Mengakses satu column dan menghasilkan Series

Rings_series = df['Rings']
print(f"Rings_series merupakan {type(Rings_series)}")
Rings_series
````

````python
# Mengakses satu column dan menghasilkan DataFrame

Rings_df = df[['Rings']]
print(f"Rings_df merupakan {type(Rings_df)}")
Rings_df
````

````python
# Mengakses beberapa column dan menghasilkan DataFrame

Length_Rings_df = df[['Length', 'Rings']]
print(f"Length_Rings_df merupakan {type(Length_Rings_df)}")
Length_Rings_df
````

#### Mengakses *Column* dengan *Dot Notation*

Untuk mengakses satu *column* yang menghasilkan *Series*, kita bisa menggunakan *dot notation*, asalkan nama *column* tersebut memenuhi [*Python variable naming rules*](https://peps.python.org/pep-0008/)

````python
df.Rings
````

#### Mengakses *Index* (Baris) dan Column Menggunakan Fungsi `loc` dan `iloc`

Fungi `loc` digunakan untuk mengakses data berdasarkan label dari *index* atau *column* dengan *range* inklusif. Sedangkan fungsi `iloc` digunakan untuk mengakses data berdasarkan posisi bilangan bulat *index* atau *column* (0 sampai jumlah baris - 1).

````python
# Mengakses data dengan label index dan column
df.loc[0:5, 'Length':'Rings'] # Baris 0 sampai 5, dari kolom Length hingga Rings
````

````python
df.loc[:, ['Length', 'Rings']] # Seluruh baris, kolom Length dan Rings saja
````

````python
# Mengakses data dengan posisi bilangan bulat index dan column
df.iloc[0:6, 1:9] # Dari baris 0 sampai 5, dari kolom Length sampai Rings
````

````python
df.iloc[:, [1, 8]] # Seluruh baris, kolom Length dan Rings saja
````

### Menangani *Missing Values*

Pada dataset, terkadang kita menemukan data yang tidak lengkap yang umumnya ditandai dengan `null`, `NA`, atau `NaN`. Hal ini dapat disebabkan oleh berbagai alasan, seperti data yang tidak tersedia walaupun sebenarnya ada atau memang tidak ada sama sekali. Untuk menangani nilai yang hilang, kita bisa menggunakan berbagai fungsi dalam Pandas.

**Mengidentifikasi *Missing Values***

Pertama-tama, mari membuka file `player_playing_time.csv` yang ada di SCeLE. Dataset ini diambil dari [🏆 UEFA EURO 2024 Database](https://www.kaggle.com/datasets/mexwell/uefa-euro-2024-database) yang merupakan kumpulan data dari pemain-pemain yang berlaga pada turnamen UEFA Euro 2024 kemarin.



````python
df2 = pd.read_csv("player_playing_time.csv")

# Menampilkan jumlah missing value per kolom
df2.isna().sum()
````

````python
# Menampilkan data dengan missing values
df2[df2.isna().any(axis=1)]
````

Ada beberapa pendekatan untuk menangani _missing values_, berikut pendekatan yang umum digunakan:
* Mengisi _missing values_ dengan nilai tertentu menggunakan _method_ <code>fillna()</code>
* Menghapus (_drop_) baris atau kolom yang mengandung _missing values_. Gunakan <code>dropna()</code> untuk menghapus __baris__ yang mengandung _missing values_, dan <code>dropna(axis='columns')</code> untuk menghapus __kolom__ yang mengandung missing values. Setelah drop, jangan lupa untuk mereset index agar tetap terurut.

````python
# Mengisi missing values dengan nilai modus

df2_fillna = df2.copy()
df2_fillna = df2_fillna.fillna(df2_fillna.mode().iloc[0])
df2_fillna.isna().sum()
````

````python
# Menghapus baris yang mengandung missing values

df2_drop_row = df2.copy()
df2_drop_row = df2_drop_row.dropna().reset_index(drop=True)
print(f"Jumlah baris tersisa adalah: {df2_drop_row.shape[0]}")
df2_drop_row.isna().sum()
````

# **NumPy**

[***NumPy***](https://numpy.org/) adalah *library* Python yang umum digunakan untuk melakukan komputasi saintifik, pengolahan, serta analisis data numerik karena performa dan kemudahannya. *NumPy* menyediakan fungsi-fungsi *high-level* untuk melakukan berbagai perhitungan secara efisien, seperti penjumlahan, pengurutan, dan sebagainya pada struktur data yang disebut ***array***.

## Array NumPy

Secara umum, *array* pada NumPy merupakan struktur data yang mirip dengan List, tetapi memiliki kemampuan yang lebih kuat dan efisien. *Array* bisa memiliki lebih dari satu dimensi, yang mana elemennya bisa berarti sebuah *array* juga, yang bisa memiliki elemen *array* juga, dan begitupun seterusnya sehingga memungkinkan penyimpanan struktur data dengan dimensi yang kompleks.

Untuk membuat objek *array*, kita bisa menggunakan fungsi `array(n)` dengan `n` berupa koleksi seperti list, tuple, dan sebagainya seperti berikut:

````python
arr = np.array([1, 2, 3, 4, 5])
arr
````

````python
print(type(arr))
````

Selain itu, terdapat fungsi bawaan untuk membuat objek *array* yang dapat digunakan sesuai kebutuhan.

````python
# Membuat array dengan nilai 0 sebanyak 5
print("np.zeros(5) : ", np.zeros(5))

# Membuat array dengan nilai 1 sebanyak 5
print("np.ones(5) : ", np.ones(5))

# Membuat array dengan nilai dalam range 1 sampai 10 dengan step sebesar 2
print("np.arange(1,10,2) : ", np.arange(1,10,2))

# Membuat array dengan nilai dalam range 1 sampai 10 sebanyak 4 dengan interval yang sama
print("np.linspace(1,10,4) : ", np.linspace(1,10,4))
````

Untuk membuat *array* multidimensi, kita dapat menyediakan *collection* sebagai berikut:

### Nomor 2

Terdapat beberapa *array* multidimensi sebagai berikut:

````python
A = np.array([[1, 4, 3, 2], [0, 2, 1, 4], [3, 0, 2, 1], [5, 1, 0, 2]])
B = np.array([[2, 1, 3, 4], [0, 3, 4, 1], [4, 2, 1, 0], [1, 3, 2, 5]])
C = np.array([[0, 3, 1, 4], [2, 0, 3, 1], [1, 2, 4, 0], [5, 1, 2, 3]])
D = np.array([[4, 2, 0, 1], [3, 1, 2, 0], [1, 4, 2, 3], [0, 5, 4, 1]])
````

Dari ketiga multidimensi array di atas, lakukanlah langkah-langkah berikut ini.
1. Cari absolute dari penjumlahan antara A dan C yang kemudian dikurangi hasil penjumlahan B dan D
2. Buat matrix baru yang isinya adalah hasil akar (square root) tiap elemen dari matrix pada langkah 1 yang setiap elemennya dikali 2
3. Jalankan operasi invers pada matrix pada langkah 2
4. Terapkan operasi dot product pada matrix pada langkah 3 dengan matrix B
5. Jalankan perkalian skalar pada matrix A dengan hasil dot product pada langkah 4
*hint*: anda dapat menggunakan fungsi `sqrt()`

Secara matematis, langkah di atas dapat dituliskan sebagai berikut.
1. $V = |A + C - (B + D)|$
2. $W =$ sqrt(sebuah matrix yang tiap elemennya adalah elemen V) * 2
3. $X = W^{-1}$
4. $K = X\cdot{B}$
5. $result = K*A$

````python
# Jawaban nomor 2
V = np.abs(A + C - (B + D))

W = np.sqrt(V) * 2

X = np.linalg.inv(W)

K = np.dot(X, B)

result = K * A

result
````
